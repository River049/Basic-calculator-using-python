```python
a = int(input('please enter the first number: '))
b = int(input('please enter the second number: '))
print('thank you so much for entering the two number.')
oper = input('please enter one operator (+, -, *, /, //, %, **): ')
if (oper == '+'):
    print(a + b)
elif(oper == '-'):
    print(a - b)
elif(oper == '*'):
    print(a * b)
elif(oper == '/'):
    print(a / b)
elif(oper == '//'):
    print(a // b)
elif(oper == '%'):
    print(a % b)
elif(oper == '**'):
    print(a ** b)
print('thank you for using my calculator')
```
